```python
import pandas as pd
import numpy as np

dfMath = pd.read_csv("Main Datasets/Maths.csv")
dfPor = pd.read_csv("Main Datasets/Portuguese.csv")
```

```python
df = pd.concat([dfMath, dfPor])
```

```python
df1 = pd.DataFrame()
df1["School"] = df["school"].replace(to_replace = ("GP","MS"), value = ("Gabriel Pereira", "Mousinho da Silveira"))
df1["Sex"] = df["sex"].replace(to_replace = ("F","M"), value = ("Female", "Male"))
df1["Age"] = df["age"].replace(to_replace = (21,22), value = ("20+"))
df1["Area Type"] = df["address"].replace(to_replace = ("U","R"), value = ("Urban", "Rural"))
df1["Family Size"] = df["famsize"].replace(to_replace = ("LE3","GT3"), value = ("Less or Equal to 3", "More than 3"))
df1["Parents Cohabitation"] = df["Pstatus"].replace(to_replace = ("T","A"), value = ("Together", "Apart"))
df1["Mother's Education"] = df["Medu"].replace(to_replace = (0,1,2,3,4), value = ("None", "Primary","5th to 9th grade","Secondary","Higher"))
df1["Father's Education"] = df["Fedu"].replace(to_replace = (0,1,2,3,4), value = ("None", "Primary","5th to 9th grade","Secondary","Higher"))
df1["Mother's Job"] = df["Mjob"].replace(to_replace = ("teacher","health","services","at_home"), value = ("Teacher","Health","Services","At Home"))
df1["Fathers's Job"] = df["Fjob"].replace(to_replace = ("teacher","health","services","at_home","other"), value = ("Teacher","Health","Services","At Home","Other"))
df1["Reason"] = df["reason"].replace(to_replace = ("home","reputation","course","other"), value = ("Close to Home","School Reputation","Course Preference","Other"))
df1["Guardian"] = df["guardian"].replace(to_replace = ("father","mother","other"), value = ("Father","Mother","Other"))
df1["Study Time"] = df["studytime"].replace(to_replace = (1,2,3,4), value = ("Less than 2 hours","2 to 5 hours","5 to 10 hours", "More than 10 hours"))
df1["Failed"] = df["failures"]
df1["Extra School Support"] = df["schoolsup"]
df1["Extra Paid Courses"] = df["paid"]
df1["Extra Activities"] = df["activities"]
df1["Attended Nursery School"] = df["nursery"]
df1["Wants Higher Education"] = df["higher"]
df1["Internet Access"] = df["internet"]
df1["In a Romantic Relation"] = df["romantic"]
df1["Family Relation"] = df["famrel"].replace(to_replace = (1,2,3,4,5), value = ("Very Bad","Bad","Normal","Good","Excellent"))
df1["Free Time"] = df["freetime"].replace(to_replace = (1,2,3,4,5), value = ("Very Low","Low","Normal","High","Very High"))
df1["Go Out"] = df["goout"].replace(to_replace = (1,2,3,4,5), value = ("Very Low","Low","Normal","High","Very High"))
df1["Workday Alcohol Consumption"] = df["Dalc"].replace(to_replace = (1,2,3,4,5), value = ("Very Low","Low","Normal","High","Very High"))
df1["Weekly Alcohol Consumption"] = df["Walc"].replace(to_replace = (1,2,3,4,5), value = ("Very Low","Low","Normal","High","Very High"))
df1["Health Status"] = df["health"].replace(to_replace = (1,2,3,4,5), value = ("Very Low","Low","Normal","High","Very High"))
df1["# of Class Absent"] = df["absences"]
```

```python
df1.to_csv("Merged - Alcohol Effects on Study.csv", index=False)
```

```python
FamilyInfo = df1[[
    "Family Size",
    "Parents Cohabitation",
    "Mother's Education",
    "Father's Education",
    "Mother's Job",
    "Fathers's Job",
    "Family Relation",
    "Guardian"
]]

FamilyInfo.drop_duplicates( inplace=True, ignore_index = True)
FamilyInfo["FamilyInfoSK"] = range(1, len(FamilyInfo)+1)

FamilyInfo.to_csv("Data Modeled Tables/Family Info.csv", index = False)
```

```python
StudentInfo = df1[[
    "Extra School Support",
    "Extra Paid Courses",
    "Extra Activities",
    "Attended Nursery School",
    "Wants Higher Education",
    "Internet Access",
    "In a Romantic Relation",
    "Free Time",
    "Go Out"
]]

StudentInfo.drop_duplicates( inplace=True, ignore_index = True)
StudentInfo["StudentInfoSK"] = range(1, len(StudentInfo)+1)

StudentInfo.to_csv("Data Modeled Tables/Student Info.csv", index = False)
```

```python
df1= df1.merge(FamilyInfo, how = "left", on = [
    "Family Size",
    "Parents Cohabitation",
    "Mother's Education",
    "Father's Education",
    "Mother's Job",
    "Fathers's Job",
    "Family Relation",
    "Guardian"])

df1= df1.merge(StudentInfo, how = "left", on = [
    "Extra School Support",
    "Extra Paid Courses",
    "Extra Activities",
    "Attended Nursery School",
    "Wants Higher Education",
    "Internet Access",
    "In a Romantic Relation",
    "Free Time",
    "Go Out"])
```

```python
Fact = df1[[
    "School",
    "Sex",
    "Age",
    "Area Type",
    "Reason",
    "Study Time",
    "Failed",
    "Workday Alcohol Consumption",
    "Weekly Alcohol Consumption",
    "Health Status",
    "# of Class Absent",
    "FamilyInfoSK",
    "StudentInfoSK"
]]

Fact.to_csv("Data Modeled Tables/Fact.csv", index = False)
```
