```python
import pandas as pd

df = pd.read_csv("owid-covid-data.csv")
```

```python
Location = df[["iso_code","continent","location"]]
Location.drop_duplicates(inplace = True, ignore_index= True)

Location.to_csv("Covid - Split Tables/Locations.csv",index= False, columns=["iso_code","continent","location"])
```

```python
Demographic = df[["population",
"population_density",
"median_age",
"aged_65_older",
"aged_70_older",
"gdp_per_capita",
"extreme_poverty"]]


Demographic.drop_duplicates(inplace = True, ignore_index= True)
Demographic['Demographic ID'] = range(1, len(Demographic) + 1)
Demographic.to_csv("Covid - Split Tables/Demographic.csv",index= False)
```

```python
Vaccination = df[[
    "new_tests",
    "positive_rate",
    "tests_per_case",
    "tests_units",
    "people_vaccinated",
    "people_fully_vaccinated",
    "total_boosters",
    "new_vaccinations",
    "icu_patients",
    "hosp_patients"]]

Vaccination.drop_duplicates(inplace = True, ignore_index= True)
Vaccination['Vaccination ID'] = range(1, len(Vaccination) + 1)

Vaccination.to_csv("Covid - Split Tables/Vaccination.csv",index= False)
```

```python
Other = df[[
    'stringency_index',
    'cardiovasc_death_rate',
    'diabetes_prevalence',
    'female_smokers',
    'male_smokers',
    'handwashing_facilities',
    'hospital_beds_per_thousand',
    'life_expectancy',
    'human_development_index',
    'excess_mortality_cumulative_absolute',
    'excess_mortality_cumulative',
    'excess_mortality']]

Other.drop_duplicates(inplace = True, ignore_index= True)

Other['Prevention ID'] = range(1, len(Other) + 1)
Other.to_csv("Covid - Split Tables/Other.csv",index= False)
```

```python
fact = df
fact = fact.merge(Demographic, how = 'left', on = ["population","population_density","median_age","aged_65_older","aged_70_older","gdp_per_capita","extreme_poverty"])
fact = fact.merge(Other, how = 'left', on = ['stringency_index','cardiovasc_death_rate','diabetes_prevalence','female_smokers','male_smokers','handwashing_facilities','hospital_beds_per_thousand','life_expectancy','human_development_index','excess_mortality_cumulative_absolute','excess_mortality_cumulative','excess_mortality'])
fact

# factVaccination = df
# factVaccination = factVaccination.merge(Demographic, how = 'left', on = ["population","population_density","median_age","aged_65_older","aged_70_older","gdp_per_capita","extreme_poverty"])
# factVaccination = factVaccination.merge(Other, how = 'left', on = ['stringency_index','cardiovasc_death_rate','diabetes_prevalence','female_smokers','male_smokers','handwashing_facilities','hospital_beds_per_thousand','life_expectancy','human_development_index','excess_mortality_cumulative_absolute','excess_mortality_cumulative','excess_mortality'])
# factVaccination
```
